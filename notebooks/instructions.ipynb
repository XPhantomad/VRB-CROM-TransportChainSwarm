4 Robots deliver load objects (black cylinders) from the prey to the nest collaboratively. They handover their load to form a chain with one head, one tail and two intermediates. The approach is completely decentralized and uses neither global communication nor centralized control. Communication is established via colored LEDs on each robot. The chain is daynamic thus if one robot fails the other robots overtake its role and form a wokring transport chain again. The Webapp visualized the currently exposed Message, Role, Team and some other useful parameters of each robot.
